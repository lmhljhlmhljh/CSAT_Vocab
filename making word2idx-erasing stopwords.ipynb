```python
import pandas as pd
import numpy as np
import os
import re, nltk
```

```python
# Loading word2idx
import json
with open('raw_text.json') as json_file:
    raw_text = json.load(json_file)
```

```python
len(raw_text)
```

```python
morph = nltk.word_tokenize(raw_text)
print(type(morph))
print(len(morph))
```

# Erasing stopwords from NLTK

```python
# open english_stopword.txt
f = open('english_stopword.txt', 'r', encoding = 'utf-8')
if f.mode == 'r':
    content =f.read()

content = content.split('\n')
content
```

```python
num = 0
new_morph = []
for word in morph :
    if word.lower() not in content :
        new_morph.append(word.lower())
        num += 1
count = len(new_morph)
print(count)
new_morph[0:20]
```

# Making word_count
- count the each morph

```python
word_count = {}
```

```python
for word in new_morph :
    if word not in word_count :
        word_count[word] = 0
    else :
        word_count[word] += 1

print(len(word_count))
print(count)
```

```python
# ratio를 위한 코드
w_count =  {k:v/count*100 for (k,v) in word_count.items()}
```

# Making word2idx
- 내림차순 정렬
- ratio of morphs

```python
# word_count ==> new_word_count
new_word_count = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
new_word_count_dict =  {k:v for (k,v) in new_word_count}

# 내림차순으로 사전 생성
word2idx = {}
num = 0
for keys in new_word_count_dict :
    num += 1
    word2idx[keys] = num
```

```python
print(len(word2idx))
word2idx
```

```python
# ratio of morphs
# 반올림 함수 (round(num,자릿수))
new_count = sorted(w_count.items(), key=lambda x: x[1], reverse=True)
new_count_dict =  {k:round(v,4) for (k,v) in new_count}
new_count_dict
```

# Making csv files

```python
df = pd.DataFrame(list(word2idx.items()), columns=['단어', '순서'])
df_percent = pd.DataFrame(list(new_count_dict.items()), columns=['단어', '비율'])
df['비율(%)'] = df_percent.비율
print(len(df))
df.head(10)
```

```python
# Saving CSV

df.to_csv('Word Cloud_최종(no stopwords).csv', index = False, encoding = 'utf-8')
```
